## I) Cleaning raw JSON tweets data scraped using snscrape library

### Importing required libraries

```python
# Importing required libraries

import pandas as pd
from collections import Counter
import re
import heapq
from pandas.io.json import json_normalize
import warnings
warnings.filterwarnings("ignore")
```

### Read raw JSON tweets data

```python
# Read JSON file containing tweets data and removce tweets not in English

raw_tweets = pd.read_json(r'archive/farmers-protest-tweets-2021-03-5.json', lines=True)
raw_tweets = raw_tweets[raw_tweets['lang']=='en']
print("Shape: ", raw_tweets.shape)
raw_tweets.head(5)
```

### Normalize 'user' field in raw_tweets

```python
# Normalize 'user' field

users = json_normalize(raw_tweets['user'])
users.drop(['description', 'linkTcourl'], axis=1, inplace=True)
users.rename(columns={'id':'userId', 'url':'profileUrl'}, inplace=True)
users.head(5)
```

### Create users DF

```python
# Create DataFrame and remove duplicates

users = pd.DataFrame(users)
users.drop_duplicates(subset=['userId'], inplace=True)
print("Shape: ", users.shape)
users.head(5)
```

### Create tweets DF

```python
# Transform 'raw_tweets' DataFrame

# Add column for 'userId'
user_id = []
for user in raw_tweets['user']:
    uid = user['id']
    user_id.append(uid)
raw_tweets['userId'] = user_id

# Remove less important columns
cols = ['url', 'date', 'renderedContent', 'id', 'userId', 'replyCount', 'retweetCount', 'likeCount', 'quoteCount', 'source', 'media', 'retweetedTweet', 'quotedTweet', 'mentionedUsers']
tweets = raw_tweets[cols]
tweets.rename(columns={'id':'tweetId', 'url':'tweetUrl'}, inplace=True)
tweets.head(5)
```

```python
# Convert to DataFrame, remove duplicates and keep only English tweets

tweets = pd.DataFrame(tweets)
tweets.drop_duplicates(subset=['tweetId'], inplace=True)
print("Shape: ", tweets.shape)
tweets.head(5)
```

## II) GitFlow diagnostic test

### 1) Top 10 tweets function

```python
def top_10_tweets():
    top_tweets = tweets.sort_values(by="retweetCount", ascending=False)
    return top_tweets.head(10)
```

### 2) Top 10 users function

```python
def top_10_users():
    tweets_users = tweets.groupby(by="userId").size().reset_index(name='counts')
    top_10_users_id = tweets_users.sort_values(by="counts", ascending=False).head(10)
    top_10_users = pd.merge(top_10_users_id, users, on=["userId"])
    return top_10_users
```

### 3) Top 10 days function

```python
def top_10_days():
    tweets_days = tweets.groupby(by="date").size().reset_index(name='counts')
    top_10_days = tweets_days.sort_values(by="counts", ascending=False).head(10)
    return top_10_days
```

### 4) Top 10 hashtags function

```python
def top_10_hashtags():
    renderedContent_column = tweets['renderedContent'].tolist()
    hash_counts = Counter(re.findall(r'#([a-z0-9]+)', ' '.join(renderedContent_column), re.I))
    # print(dict(hash_counts))
    return heapq.nlargest(10, hash_counts, key=hash_counts.get)
```

### Main function

```python
def main(type):
    if type == 1:
        return top_10_tweets()
    elif type == 2:
        return top_10_users()
    elif type == 3:
        return top_10_days()
    elif type == 4:
        return top_10_hashtags()
    else:
        print("Input incorrecto")
```
